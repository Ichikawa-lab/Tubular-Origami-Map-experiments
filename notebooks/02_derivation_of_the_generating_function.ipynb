# Reproducing the Results of Eq. (A17)

This notebook reproduces the derivation of the integral expression (Eq A.17) of the generating function presented in the paper.

paper: Invariant Curves and the Variational Structure in Tubular Origami Dynamical Systems (https://arxiv.org/abs/2603.02410)

## Setup

First, we define the symbols and the functions such as A and B according to the notation used in the paper.

```python
import sympy as sp

# 1. Define symbolic variables
I, theta, rho = sp.symbols('I theta rho', real=True, positive=True)
alpha, beta, l_L, l_R = sp.symbols('alpha beta l_L l_R', real=True, positive=True)
l = sp.symbols('l', real=True, positive=True)

# Radii r1, r2, r3 (assumed to be positive real numbers)
r_1, r_2, r_3 = sp.symbols('r1 r2 r3', positive=True)

# Mountain/Valley (MV) sign: +1 or -1
sigma = sp.symbols('sigma', real=True)

# -------------------------------------------------------------- #
# Definitions of geometric parameters appearing in the calculation

alpha = (l_L**2 + (2*I)**2 - l_R**2)/(2*l_L*(2*I))
beta = sp.sqrt(1 - alpha**2)

e1 = (r_1**2 + l_L**2 - r_2**2) / (2*l_L)
e2 = (r_1**2 - r_3**2 + 4*I**2 - 4*I*e1*alpha) / (4*I*beta)

gamma = r_1**2 - e1**2 - e2**2
e3 = sigma * sp.sqrt(gamma)

# Coefficients used in the integrand
A = -e1 * beta + e2 * alpha
B = e3
# -------------------------------------------------------------- #
```

## Verification of the Expression inside $\beta\sqrt{\gamma}$

In the calculation presented in the paper, the following expression (Eq. (69)) is obtained:

$ \psi(I) = \frac{I(B^{\prime}A-A^{\prime}B)}{A^2+B^2} = \frac{\sigma}{\beta\sqrt{\gamma}} R(I). $

Here we compute the expression inside \(\beta\sqrt{\gamma}\), which corresponds to the irrational part of the integrand.  
The result can be verified using the code below.

```python
# Verification: expression inside the square root multiplied by beta^2

# Asymmetric case
expr_asym = (beta**2 * gamma).simplify()

# Symmetric case (l_L = l_R = l, r_1 = r_3 = r)
expr_sym = (beta**2 * gamma).subs([
    (l_L, l),
    (l_R, l),
    (r_1, r_1),
    (r_3, r_1)
]).simplify()

# Display both results
print("Asymmetric case:")
display(expr_asym)

print("Symmetric case:")
display(expr_sym)
```

For the rational part $R(I)$, the result below confirms that it is indeed a rational function.

```python
# ---------------------- Definition of rational functions ---------------------- #

# Rational expression corresponding to A
R_A = -e1 * beta**2 + alpha * (r_1**2 - r_3**2 + 4*I**2 - 4*e1*alpha*I) / (4*I)

# Derivative of R_A with respect to I
R_A_diff = sp.diff(R_A, I)

R_diff_A = R_A_diff + alpha * sp.diff(alpha, I) * R_A / (beta**2)

# Rational part of the integrand
R = I * (sp.diff(gamma, I) / 2 * R_A - R_diff_A * gamma) / (A**2 + B**2)

# ---------------------- Symmetric case evaluation ---------------------- #

# Substitute symmetric conditions: l_L = l_R = l, r_1 = r_3 = r
R_sym = R.subs([
    (l_L, l),
    (l_R, l),
    (r_1, r_1),
    (r_3, r_1)
]).simplify()

# Display the result for the symmetric system
print("R(I)=")
display(R_sym)
```

## Partial Fraction Decomposition in the Symmetric Case

```python
# Define auxiliary variables used in the substitution
K, t = sp.symbols('K t', real=True)

# Note: the outer factor (1 + t^2) is not included here

# Substitute the symmetric conditions
integral_term = R.subs([
    (l_L, l),
    (l_R, l),
    (r_1, r_1),
    (r_3, r_1)
]).simplify()

# Apply the change of variables I = 2Kt / (1 + t^2)
integral_term = integral_term.subs(I, 2*K*t/(1+t**2))

# Use sigma^2 = 1
integral_term = integral_term.subs(sigma**2, 1).simplify()

# Multiply by the Jacobian factor from the substitution
integral_term = 2 * integral_term / (1 + t**2)

# Extract numerator and denominator
numerator, denominator = sp.fraction(integral_term)

# Factor both numerator and denominator with respect to t
numerator = sp.factor(sp.sympify(numerator), t)
denominator = sp.factor(sp.sympify(denominator), t)

# Simplify the rational expression
f = (numerator / denominator).simplify()
print("Rational expression after substitution and simplification:")
display(f)

# Perform partial fraction decomposition with respect to t
result = sp.apart(f, t)
print("Partial fraction decomposition with respect to t:")
display(result)
```

# Integration Involving arctan

The integrals of the arctangent-type terms are:

$$
\int \frac{1}{l t^2 + 2 K t + l} \, dt
= \frac{1}{\sqrt{l^2 - K^2}} \arctan \left( \frac{l t + K}{\sqrt{l^2 - K^2}} \right) + C
$$

$$
\int \frac{1}{l t^2 - 2 K t + l} \, dt
= \frac{1}{\sqrt{l^2 - K^2}} \arctan \left( \frac{l t - K}{\sqrt{l^2 - K^2}} \right) + C
$$
